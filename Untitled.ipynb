```python
import pandas as pd

# Load the dataset
df = pd.read_csv('taxis.csv')

# View first 5 rows
print("üìÑ First 5 rows:")
print(df.head())

# Shape of the dataset
print("\nüìê Shape of dataset (rows, columns):", df.shape)

# Data types
print("\nüîç Data types:")
print(df.dtypes)

# Check missing values
print("\n‚ùì Missing values per column:")
print(df.isnull().sum())

# Check duplicates
print("\nüìã Number of duplicate rows:", df.duplicated().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert pickup and dropoff to datetime
df['pickup'] = pd.to_datetime(df['pickup'])
df['dropoff'] = pd.to_datetime(df['dropoff'])

# Summary statistics
print("üìä Summary statistics:")
print(df.describe())

# Plot histograms for numerical features
numerical_cols = ['passengers', 'distance', 'fare', 'tip', 'tolls', 'total']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i + 1)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# Categorical columns
categorical_cols = ['color', 'payment', 'pickup_zone', 'dropoff_zone', 'pickup_borough', 'dropoff_borough']

# Count unique values and plot distributions
for col in categorical_cols:
    print(f"\nüü© {col.upper()} - Unique values: {df[col].nunique()}")
    print(df[col].value_counts(dropna=False).head(10))  # top 10 categories

    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index[:10])
    plt.title(f'Distribution of {col} (Top 10)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Select only numerical columns
numerical_cols = ['passengers', 'distance', 'fare', 'tip', 'tolls', 'total']

# Correlation matrix
corr = df[numerical_cols].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='distance', y='fare', hue='passengers', palette='viridis', alpha=0.7)
plt.title('Fare vs Distance by Passenger Count')
plt.xlabel('Distance (miles)')
plt.ylabel('Fare ($)')
plt.legend(title='Passengers')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(data=df, x='payment', y='tip')
plt.title('Tip Distribution by Payment Method')
plt.xticks(rotation=45)
plt.show()
```

```python
# Extract hour from pickup time
df['pickup_hour'] = df['pickup'].dt.hour

# Group and plot
hourly_tip = df.groupby('pickup_hour')['tip'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.lineplot(data=hourly_tip, x='pickup_hour', y='tip', marker='o')
plt.title('Average Tip by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Tip ($)')
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

```python
# Number of rides per hour
rides_per_hour = df['pickup_hour'].value_counts().sort_index()

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=rides_per_hour.index, y=rides_per_hour.values, palette='crest')
plt.title('Number of Rides per Hour')
plt.xlabel('Hour of Day (0 = Midnight)')
plt.ylabel('Number of Rides')
plt.grid(True)
plt.show()
```

```python
# Count rides by day
rides_per_day = df['pickup_day'].value_counts().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=rides_per_day.index, y=rides_per_day.values, palette='Set2')
plt.title('Number of Rides per Day of Week')
plt.xlabel('Day')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by day and find average fare
fare_by_day = df.groupby('pickup_day')['fare'].mean().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=fare_by_day.index, y=fare_by_day.values, palette='pastel')
plt.title('Average Fare by Day of Week')
plt.xlabel('Day')
plt.ylabel('Average Fare ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Columns to check
outlier_cols = ['distance', 'fare', 'tip', 'total']

# Plot boxplots
plt.figure(figsize=(14, 8))
for i, col in enumerate(outlier_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
# Show top 5 largest values in each key column
for col in outlier_cols:
    print(f"\nTop 5 highest {col}:")
    print(df[[col]].sort_values(by=col, ascending=False).head(5))
```
